```python
fig,axs = plt.subplots(1,2,figsize=(15,7))
plt.subplot(121)
sns.histplot(data=df,x="RUL",bins=30,kde=True,color='g')
plt.subplot(122)
sns.histplot(data=df,x="RUL",kde=True,hue='Temperature (°C)')
plt.show()
```

```python
fig,axs = plt.subplots(1,2,figsize=(15,7))
plt.subplot(121)
sns.histplot(data=df,x="RUL",bins=30,kde=True,color='g')
plt.subplot(122)
sns.histplot(data=df,x="RUL",kde=True,hue='Temperature (°C)')
plt.show()
```

```python
plt.figure(figsize=(18,8))
plt.subplot(1, 4, 1)
plt.title('Max. Voltage Dischar. (V)')
sns.violinplot(y='Max. Voltage Dischar. (V)',data=df,color='red',linewidth=3)
plt.subplot(1, 4, 2)
plt.title('Time constant current (s)')
sns.violinplot(y='Time constant current (s)',data=df,color='green',linewidth=3)
plt.subplot(1, 4, 3)
plt.title('Temperature (°C)')
sns.violinplot(y='Temperature (°C)',data=df,color='blue',linewidth=3)
plt.show()
```

```python
plt.subplots(1,4,figsize=(16,5))
plt.subplot(141)
sns.boxplot(df['Max. Voltage Dischar. (V)'],color='skyblue')
plt.subplot(142)
sns.boxplot(df['Time constant current (s)'],color='hotpink')
plt.subplot(143)
sns.boxplot(df['Temperature (°C)'],color='yellow')
plt.subplot(144)
sns.boxplot(df['RUL'],color='lightgreen')
plt.show()
```
